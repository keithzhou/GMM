$$ P(x) = \sum_{k=1}^{K} \pi_k \prod_{i=1}^{D} \mu_{ki}^{x_i} (1-\mu_{ki})^{1-x_i}$$

model parameters: $\pi_j, \mu_{ji}$

Adding hidden variables allows complicated marginal distribution over observed variabled to be assembled from simpler joint distributions:

  1. $P(x|z=j) = \prod_{i=1}^{D} \mu_{ji}^{x_i} (1-\mu_{ji})^{1-x_i}$
  2. $P(x,z) = \pi_z p(x|z)$

Note $ln(P(x,z))$ is simpler than $ln(P(x))$

```python
# data
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

SHAPEIMAGE = (28,28)
DATA = np.fromfile("digits-ubyte",dtype=np.uint8)[16:].reshape(60000,28*28)
DATA = np.where(DATA < 128, 0.0, 1.0)
DATA = DATA[np.random.permutation(len(DATA))][:6000] # use only 6000 data points
```

```python
# an example of the input dataset
for i in range(5):
    plt.imshow(DATA[np.random.randint(len(DATA))].reshape(SHAPEIMAGE), cmap='Greys')
    plt.show()
```

For EM:
$$Q=\delta(\theta-\hat\theta)Q(h)$$

$$F=-H_\sigma - ln(P(\hat\theta)) + \sum_t \sum_{h^(t)}Q(h^{(t)})ln\frac{Q(h^{(t)})}{P(h^{(t)},v^{(t)}|\hat\theta}$$

Step E: $Q(h^{(t)}) = P(h^{(t)}|v^{(t)},\hat\theta)$
  
  * $$Q^{(t)}(z) = P(z|x^{(t)},\hat\theta) = \frac{P(z,x^{(t)}|\hat\theta)}{\sum_z P(z,x^{(t)}|\hat\theta)} = \frac{\pi_z \prod_i \mu_{i}^ {x^{(t)}} (1-\pi_z) ^ {1-x^{(t)}}}{\sum_z\pi_z \prod_i \mu_{i}^ {x^{(t)}} (1-\pi_z) ^ {1-x^{(t)}}} $$

Step M: $\theta = \arg\max_\theta \sum_t\sum_{h^{(t)}} Q(h^{(t)})ln(P(h^{(t)},v^{(t)}|\theta))$

  * $$\pi_j = \frac{Q(z^{(t)}=j)}{T}$$
  * $$\mu_{ji} = \frac{\sum_t Q(z^{(t)}=j)x_{i}^{(t)}}{\sum_t Q(z^{(t)}=j)}$$

```python
# E
def doE():
    Q_z_log = np.log(pi) + np.sum(DATA[:,np.newaxis,:] * np.log(mu) + (1 - DATA[:,np.newaxis,:]) * np.log(1-mu), axis=2)
    Q_z = np.exp(Q_z_log - np.max(Q_z_log,axis=1)[:,np.newaxis])
    Q_z /= np.sum(Q_z, axis=1)[:, np.newaxis]
    return Q_z

#Q = doE()
```

```python
def doM(Q):
    piN = np.sum(Q, axis=0) / len(Q)
    muN = np.sum(Q[:,:, np.newaxis] * DATA[:,np.newaxis,:], axis=0) / np.sum(Q,axis=0)[:, np.newaxis]
    return (piN,muN)
```

```python
K = 20
pi = np.ones(K) * (1.0/K)
mu = np.random.random((K,len(DATA[0])))

for i in range(0,101):
    if i % 50 == 0:
        for j in range(len(mu)):
            plt.imshow(mu[j].reshape(SHAPEIMAGE), cmap='Greys')
            plt.show()
    Q = doE()
    print "iteration:", i, "free energy:", calculateF(Q)   
    pi,mu = doM(Q)
    
    # avoid numerical underflow
    pi[pi < 1e-6] = 1e-6
    pi[pi > 1.0 - 1e-6] = 1.0 - 1e-6
    mu[mu < 1e-6] = 1e-6
    mu[mu > 1.0 - 1e-6] = 1.0 - 1e-6
```

```python
def calculateF(Q):
    part1 = np.sum(Q * np.log(Q + 1e-6))
    part2 = np.sum(Q * np.log(pi)[np.newaxis,:])
    part3 = np.sum(Q * np.sum(DATA[:,np.newaxis,:] * np.log(mu)[np.newaxis,:,:] + (1-DATA)[:,np.newaxis,:] * np.log(1-mu)[np.newaxis,:,:],axis=2))
    return part1 - part2 - part3
```
